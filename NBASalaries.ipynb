# In the NBA, is there an effect of being a forward on your salary?

```r
install.packages("wooldridge")
library(wooldridge)
library(ggplot2)
library(tidyverse)
```

```r
nbasal
```

```r
summary(nbasal)
```

```r
model <- lm(lwage ~ forward, data = nbasal)

summary(model)
```

## We can see here that being a forward increases wage by approximatelty 17.56% and this is not statistically significant because it is less than two times the standard error.

## However, we believe there are a lot of confounding variables so let's take this opprtunity to measure for omitted variable bias with a few variables

## 1. exper - how many years the player has been a professional
## 2. allstar - binary variable of whether the player has been an allstar or not
## 3. games - average amount of games a player plays in a year

## __________________________________________________________________________________________________________________________________________
## Let's start with exper and run the long regression

```r
model_2 <- lm(lwage ~ forward + exper, data = nbasal)

summary(model_2)
```

## Here we see that including experience actually increases the impact of being a forward on your salary and the t value increases. However it is still statistically insignificant.

```r
model_3 <- lm(exper ~ forward, data = nbasal)

summary(model_3)
```

## It looks like being a forward decreases your amount of experience however this is not statistically significant.

## Let's caclualte OVB

```r
ovb_exper <- -0.001315 * 0.10531
ovb_exper
```

## We do not gain much from this regression. There is a lot of stastistical insignificance so it is hard to make a conclusion. Let's check out the impact of being an allstar on your salary

## Again let's start out with the long regression

```r
model_4 <- lm(lwage ~ forward + allstar, data = nbasal)

summary(model_4)
```

## Here, forward's impact on salary has increased however still statistically insignifcant. What stands out here is that being an allstar increases your salary by approximately 81.8% and this is statistically significant.

## Now let's run the auxiliary regression

```r
model_5 <- lm(allstar ~ forward, data = nbasal)

summary(model_5)
```

## Again there is no statistical significance.

## Let's try out games

```r
model_6 <- lm(lwage ~ forward + games, data = nbasal)

summary(model_6)
```
